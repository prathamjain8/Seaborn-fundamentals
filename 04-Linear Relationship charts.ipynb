```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
diamonds = pd.read_csv("D:/STUDIES/Data Visualization/Data/Diamonds Prices2022.csv")
```

```python
sns.regplot(
    x="carat",
    y="price",
    data = diamonds
);
```

```python
sns.lmplot(
    x = "carat",
    y="price",
    data = diamonds
);
```

```python
sns.lmplot(
    x = "carat",
    y="price",
    hue="clarity",
    data = diamonds
);
```

```python
sns.lmplot(
    x = "carat",
    y="price",
    row="clarity",
    col="cut",
    data = diamonds
);
```

```python
sns.jointplot(
    x = "carat",
    y = "price",
    data = diamonds
)
```

```python
sns.jointplot(
    x = "carat",
    y = "price",
    color="orange",
#     scatter_kws = {'alpha':.1},
    data = diamonds,
    kind='scatter',
    hue="clarity",
    palette = "viridis"
)
```

```python
sns.pairplot(
    diamonds.iloc[:,1:],
    corner=True,   # it is helpful in removing the redundant pairs
#     diag_kind="kde",
    hue="color",
    palette="viridis"
);
```
