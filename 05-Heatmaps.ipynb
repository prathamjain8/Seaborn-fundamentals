```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
diamonds = pd.read_csv("D:/STUDIES/Data Visualization/Data/Diamonds Prices2022.csv")

diamonds.head()
```

```python
diamonds_pivot = diamonds.pivot_table(index="cut", columns="color", values={"carat":"mean"})
```

```python
diamonds_pivot.head()
```

```python
sns.set(rc = {'figure.figsize': (15,8)})

sns.heatmap(
    diamonds_pivot.droplevel(0, axis=1),
    annot=True,
    fmt='g', 
    cmap="RdYlGn"
);
```
