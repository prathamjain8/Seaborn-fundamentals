```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
hotels = pd.read_excel("D:/STUDIES/Data Visualization/Data/HotelCustomersDataset.xlsx")

hotels.head()
```

```python
top5_countries = hotels.query("Nationality in ['GBR', 'PRT', 'FRA', 'DEU', 'ESP']")
```

```python
sns.boxplot(
    y = "LodgingRevenue",
    data = top5_countries.query("LodgingRevenue < 1500")
)
```

```python
sns.violinplot(
    x = "Nationality",
    y = "Age",
    data = top5_countries
)
```

```python
sns.violinplot(
    x = "Nationality",
    y = "LodgingRevenue",
    data = top5_countries
)
```

```python
sns.violinplot(
    x = "Nationality",
    y = "LodgingRevenue",
    data = top5_countries.query("LodgingRevenue < 1500")
)
```
