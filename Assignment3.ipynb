```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
hotels = pd.read_excel("D:/STUDIES/Data Visualization/Data/HotelCustomersDataset.xlsx")

hotels.head()
```

```python
top5_countries = hotels.query("Nationality in ['GBR', 'PRT', 'FRA', 'DEU', 'ESP']")
```

```python
sns.regplot(
    x = "RoomNights",
    y = "LodgingRevenue",
    data = top5_countries
);
```

```python
sns.regplot(
    x = "RoomNights",
    y = "LodgingRevenue",
    data = top5_countries.query("LodgingRevenue < 1500 and RoomNights < 15")
);
```

```python
sns.lmplot(
    x = "RoomNights",
    y = "LodgingRevenue",
    data = top5_countries.query("LodgingRevenue < 1500 and RoomNights < 15"),
    hue="Nationality"
);
```

```python
sns.pairplot(
    data = top5_countries.iloc[:,[2,5,6,10,11,12]]
);
```

```python
sns.lmplot(
    x = "RoomNights",
    y = "LodgingRevenue",
    col = "Nationality",
    data = top5_countries.query("LodgingRevenue < 1500 and RoomNights < 15"),
);
```
