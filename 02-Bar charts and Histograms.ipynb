```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
diamonds = pd.read_csv("D:/STUDIES/Data Visualization/Data/Diamonds Prices2022.csv")
```

```python
diamonds.head()
```

```python
# vertical bar chart
sns.barplot(
    x = 'cut',
    y = 'carat',
    data = diamonds
);
```

```python
# horizontal bar chart
sns.barplot(
    x = 'carat',
    y = 'cut',
    data = diamonds,
    ci=None
);
```

```python
sns.barplot(
    x = "price",
    y = "cut",
    data = diamonds,
    facecolor="white",
    edgecolor="black"
)
```

```python
# creating grouped bar chart

fig, ax = plt.subplots()

sns.barplot(
    x = "cut",
    y = "price",
    data = diamonds,
    hue="clarity",
    palette="husl"
);

ax.legend(bbox_to_anchor=(1.2,1));
```

```python
sns.histplot(
    x = "price",
    data = diamonds.query("clarity in ['I1','IF']"),
    hue = "clarity",
    bins = 10,
    palette='husl',
    stat = 'percent',
    kde=True,
#     binwidth=1000
);
```

```python
sns.histplot(
    x = "price",
    data = diamonds,
    y = 'clarity',
    cbar=True
);
```

```python
sns.scatterplot(
    x = "x",
    y = "price",
    data = diamonds,
    hue="carat",
    size = "carat",
    palette = 'husl'
);
```
